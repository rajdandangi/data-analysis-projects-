```python
#importing resources
import pandas as pd 
import json
```

```python
#importing data form path
with open('t20_json_files/t20_json_files/t20_wc_match_results.json') as f:
 data=json.load(f)
#conforming one element = len(data)
# creating data frame
df = pd.DataFrame(data[0]["matchSummary"])
df.head(10)
```

```python
#rename
df.rename({'scorecard' : 'match_id'},axis = 1, inplace = True)
```

```python
with open('t20_json_files/t20_json_files/t20_wc_batting_summary.json') as f :
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.tail(10)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x : 'out' if len(x) > 0 else "not out")
df_batting.head(10)
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting.head(10)
```
